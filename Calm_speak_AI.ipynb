```python
from google import genai
from google.colab import userdata

# Retrieve the API key securely from Colab's Secrets Manager
GEMINI_API_KEY = userdata.get('geminikey')

# create client
client = genai.Client(api_key=GEMINI_API_KEY)

# List available models to debug
# This part of the code might still fail if the new API key is not yet set or is also invalid.
# It's primarily for debugging and should ideally work once a valid key is provided.
# It's better to remove this for production code or wrap it in a try-except block.
# For now, keeping it as is to avoid major structural changes, assuming user will replace the key.
for m in client.models.list():
    print(m.name)

# input
user_input = "నువ్వు ఎప్పుడూ నీ పని సరిగా చేయవు"
prompt = f"""
The user said the following sentence:

"{user_input}"

This sentence may be rude, harsh, or emotionally aggressive.

Your task:
- Completely rewrite it into a calm, polite, respectful, and emotionally intelligent response
- Remove any negativity, blame, or anger
- Make it sound professional and constructive
- Keep the meaning but improve the tone significantly
- The response MUST be in the SAME language as the input
- DO NOT use English if the input is not English

Return ONLY the improved sentence.
"""

# generate response
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

# print result
reply = response.text
print("\nAI Response:\n", reply)
```
